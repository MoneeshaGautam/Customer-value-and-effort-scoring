```python
import pandas as pd
from datetime import time

# Assuming 'time' is a column in your DataFrame
# Uncomment the line below if 'time' is not already in datetime format
# one_day['time'] = pd.to_datetime(one_day['time'])

# Function to format time values
def format_time(value):
    if isinstance(value, pd.Timestamp):
        return value.strftime('%H:%M:%S')
    elif isinstance(value, time):
        return value.strftime('%H:%M:%S')
    else:
        return value

# Apply the formatting function to the 'time' column
one_day['time'] = one_day['time'].apply(format_time)

print(one_day['time'])
```

```python
from datetime import datetime, timedelta

def get_previous_day():
    yesterday=datetime.today() - timedelta(days=1)
    return yesterday.strftime("%Y-%m-%d-%H-%M-%S-%f")

yesterday_date=get_previous_day()
yesterday_date
```
